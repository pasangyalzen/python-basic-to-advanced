**Operator, Operand and Operation**

1. Operator

An operator is like a symbol or a tool that tells the computer what action to perform.

• Example: +, -, *, / are operators.

• Think of + as a “plus tool” that adds two numbers.


2. Operand

An operand is the value or thing on which the operator works.

• In 5 + 3, the numbers 5 and 3 are operands.

• They are like the “ingredients” that the operator uses.

3. Operation

An operation is the action that happens when the operator is applied to the operands.

• In 5 + 3 = 8, the addition (+) is the operation.

• It’s like the “process” of adding 5 and 3 to get 8.
